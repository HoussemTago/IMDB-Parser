```python
def get_season_episodes(seriesId,series) :
    urlwiki = f'https://www.imdb.com/title/{seriesId}/episodes?season={str(season)}'
    urlData = urlopen(urlwiki)
    wikiHtml = urlData.read()
    urlData.close()

    page_soup = soup(wikiHtml, 'html.parser')
    ep_soup = page_soup.find(class_="list detail eplist")
    epDict = {'Season':[],'Episode':[],'Title':[],'IMDB Rating':[],'Total Votes' : [],'Air Date':[]}
    for episode in ep_soup.find_all(True, {'class':['list_item even', 'list_item odd']}) :
        epInfo = episode.find(class_='info')
        epDict['Season'].append(season)
        epDict['Episode'].append(epInfo.meta['content'])
        epDict['Air Date'].append(str(epInfo.find(class_='airdate').string).strip())
        epDict['Title'].append(str(epInfo.strong.a.string))
        epDict['IMDB Rating'].append(float(epInfo.find(class_='ipl-rating-star__rating').string))
        epDict['Total Votes'].append(cleanVote(str(epInfo.find(class_='ipl-rating-star__total-votes').string)))
    return epDict
```

```python
seriesId = 'tt0903747'
season = 3
df = pd.DataFrame(get_season_episodes(seriesId,season))
```
